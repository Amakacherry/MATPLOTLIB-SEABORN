## INTRODUCTION TO MATPLOTLIB
It is one of the most widely used data visualisation library in Python. It was created by John Hunter, a neuro-Biologist.
To overcome the use of a lilimited license software,He set off to replace the proprietary software with a Matlab based version
that could be utilised by him and his teammates and that could be extended by multiple investigators. As a result, Matplotlib was developed as an ECOG visualisation tool. Just like Matlab, Matplotlib was equiped with a scripting interface, for a quick and easy generation of graphics represented by Pyplot.

Matplotlib architecture is composed of 3 main layers;

- The backend layer(FigureCanvas, Renderer, Event)
- The Artist Layer(where most of the heavy lifting happens. The appropriate programming paradign, when writing a web application server or a UI application or perhaps a script to be shared with other developers)
- The Scripting Layer(pyplot) The appropriate layer for everyday purposes. It's considered a wider scripting interface, to simplify common tasks and for quick and easy generation of graphics and plots.

The Backend Layer:
Has three built-in abstract interface classes:
1. Figure Canvass: matplotlib.backend_bases.FigureCanvas
 Encompassesthe area unto which the figure is drawn
2. Renderer: matplotlib.backend_bases.Renderer
 Knows how to draw on the figureCanvas
3. Event: matplotlib.backend_bases.Event
 Handles user inputs such as keyboard strokes and mouse clicks.
 
Artist Layer:
Comprised of one main object - Artist:
Knows how to use the Renderer to draw on canvas.

Title, lines, tick labels, and images, all correspond to individual Artist instances.

Two types of Artist objects:
1. Primitive: Line2D, Rectangle, Circle, and Text.
2. Composite: Axis, Tick, Axes, and Figure.

The top level Matplotlib object that contains and manages all the elements in a gien graphic is the FIGURE artist and the most important composite artist id the Axes because it is where most of the Matplotlib API plotting mtds are defined. 
It is important to know that each composite artist may contain other composite artists as well as primitive artists.
A Figure artist for e.g will contain an Axes artist as well as a Rectangle and a Text Artist.

## Putting The Artist Layer to Use

Let's try to generate a histogram of some data using the Artist layer:
We import the figureCanvas from the backend_agg and attch the figure artist straight. 
Note that agg is anti green geometry. It's a high performance library that produces attractive images

```python
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas # import FigureCanvas
from matplotlib.figure import Figure # import Figure artist
fig = Figure()
canvas = FigureCanvas(fig)

# create 10000 random numbers using numpy
import numpy as np
x = np.random.randn(10000)

ax = fig.add_subplot(111) # create an axes artist

ax.hist(x, 100) # generate a histogram of the 10000 numbers

# add a title to the figure and save it 
ax.set_title('Normal distribution with $\mu=0, \sigma=1$')
fig.savefig('matplotlib_histogram.png')
```

## As for the Scripting layer, It was developed for scientists who are not professional programmers.

The artist layer is syntactically heavy as it is meant for developers and not for individuas whose goal is to perform explanatory analysis of some data.

Matplotlibs sripting layer is essentially the Matplotlib.pyplot interface. Comprised mainly of pyplot, A scripting interface that is lighter than the Artist layer.

## Let's see how we can generate the same histogram of 10000 random values using the pyplot interface

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(10000)
plt.hist(x, 100)
plt.title("Normal distribution with $\mu = 0, \sigma = 1$")
plt.savefig("matplotlib_histogram.png")
plt.show()
```

**SEABORN AND REGRESSION PLOTS**

Seaborn is anotherPython visualisation library based on Matplotlib. It was built primarily to provide
a high level interface featuring attractive statistical graphics such as regression plots, box plots etc.
It makes creating plots very efficient. therefore, with seaborn you can generate plots with codes that are 5 times
less than with Matplotlib. 
Visuals that need 20 lines of code using Matplotlib to be created, with Seaborn, the number of line of code is 
is reduced by 5 fold. 

Let's look are creating regression plots:

Looking at the total immigration to Canada from 1980 to 2013. 
We create a scaler plot along with the regression line to highlight an entrance in the data.
With Seaborn, you can do all this with literally one line of code. 

First import seaborn as sns. Call the Seaborn regplot function, we tell it to use the dataframe df_total
and to plot the column year on the horizontal axis and the column total on the vertical axis. 
And the output of this one line of code is the scaler plot with the regression line and also 95% confidence interval.
You can perform other customisation on Seaborn, like changing the color parameter etc.
